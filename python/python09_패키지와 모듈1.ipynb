### 1. 패키지와 모듈

### 패키지 (Package)

특정 기능과 관련된 여러 모듈들을 하나의 상위 폴더에 넣어 놓은 것을 패키지라고 한다.  
패키지 안에 여러가지 하위폴더 가 더 존재할 수 있다.  
패키지는 라이브러리로 이해해도 된다

### 모듈 (Module)

파이썬에서는 모듈의 기능을 활용하여 코드를 분리하고 공유합니다.  
이때 여러 기능들(함수, 변수, 클래스 등)을 따로 구현하여 파이썬 파일(.py)을 모듈(module)이라고 한다.  
가끔 script라고도 부른다.  

모듈은 여러 여러 변수와 함수를 가지고 있는 집합체로 크게 표준 모듈과 외부 모듈로 나뉜다.  
파이썬에 기본적으로 내장되어 있는 모듈을 '표준 모듈'이라 하고  
다른 사람들이 만들어서 공개한 모듈을 '외부 모듈'이라 부른다.  

일반적으로 모듈을 가져오는 import구문은 코드의 가장 위에 작성한다

### 함수(Function)

하나의 기능성을 가진 코드의 집합이다.  
함수를 실행하기 위해서는 호출을 해야 한다.  
호출할 때 괄호 ( )를 사용하는데 내부에 매개변수를 넣고 최종적으로 리턴값 여부를 표시한다.

* 내장함수(Built-in function): 외부 모듈과는 달리 import를 하지 않는다. (예: int(), str() )
* 외장함수: import를 통하여 불러들인 외부 모듈 안에 있는 함수를 말한다. (예: random())
